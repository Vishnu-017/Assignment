```python
train = pd.read_csv('train1.csv', encoding='utf-8')
test = pd.read_csv('test1.csv',encoding='utf-8')
```

```python
X_train = train.drop('target', axis=1)
```

```python
X_test= test
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_clusters': [160, 165, 170, 175],  
    'init': ['k-means++', 'random'],
    'max_iter': [300, 500, 1000] 
}
```

```python
from sklearn.metrics import silhouette_score, make_scorer
silhouette_scorer = make_scorer(silhouette_score)
```

```python
from sklearn.cluster import KMeans
kmeans = GridSearchCV(KMeans(), param_grid, cv=5, scoring=silhouette_scorer)
```

```python
kmeans.fit(X_train)
```

```python
best_params = kmeans.best_params_
print("Best Hyperparameters:", best_params)
```

```python
best_kmeans = KMeans(**best_params)
best_kmeans.fit(X_train)
```

```python
test_clusters = best_kmeans.predict(X_test)
```

```python
silhouette = silhouette_score(X_test, test_clusters)
print("Silhouette Score with Best Hyperparameters:", silhouette)
```
